# Data Cleaning

#### Fill the missing values
we will fill the missing values for age. In order to fill missing values we use fillna method. For now we will fill the missing age by taking average of all age

```python
# Fill age column
```

```python


titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

#### We can verify that no more null data exist
we will examine data by isnull method which will return nothing

```python
# verify null value
titanic_data['Age'].isna().sum()
```

```python
# visualize null values
```

```python
sns.heatmap(titanic_data.isna())
```

#### Discard cabin column 

```python
# Drop cabin column
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
# verify dropped column
titanic_data.head()
```

#### Prepare data for model
change non-numeric values to numeric values or drop the column

```python
# check for the non-numeric column
titanic_data.dtypes
```

#### non-numeric values are : Name, Sex, Ticket and Embarked

```python
# convert sex column to numerical values
gender = pd.get_dummies(titanic_data['Sex'], drop_first=True)
gender
```

```python
titanic_data['Gender'] = gender
```

```python
titanic_data.head()
```

```python
# drop the columns which are not required
titanic_data.drop(['Name', 'Sex', 'Ticket', 'Embarked'], axis=1, inplace=True)
```

```python
titanic_data.head()
```

```python
# Separate Dependent and Independent variable

x = titanic_data[['PassengerId', 'Pclass','Age', 'SibSp','Parch','Fare','Gender']]
y = titanic_data['Survived']
```

```python
x
```

```python
y
```

# Data Modelling

**Building Model using Logistic Regression**

**Build the model**

```python
# split train and test
from sklearn.model_selection import train_test_split
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# import logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
# Fit logistic regression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
# predict
```

```python
predict=lr.predict(x_test)
```

# Testing

**See how our model is performing**

```python
# print confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
pd.DataFrame(confusion_matrix(y_test, predict), columns=['Predicted No', 'Predicted Yes'], index=['Actual no', 'Actual Yes'])
```

```python
# import classification report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predict))
```
